# Plot the figure for the temperature trend estimate

This notebook generates the figure for the paper. Uses the raw data `data/hawaii-TAVG-Trend.txt` and the estimated trend in `results/hawaii-trend.csv`.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# Load out custom tool for loading and processing the data
from mypackage.io import load_berkeley_earth_data
```

Set some plot parameters.

```python
plt.rc('font', size=7)
```

## Load the data and trend

Load the raw data and the estimated trend as pandas.DataFrames.

```python
data = load_berkeley_earth_data('../../data/hawaii-TAVG-Trend.txt')
trend = pd.read_csv('../../results/hawaii-trend.csv')
```

## Plot the figure and save it to EPS

Create the figure using the proper size for a single column (about 3.5 inches). 

```python
plt.figure(figsize=(3.5, 2.7))
plt.plot(data.year_decimal, data.monthly_temperature, '-', linewidth=0.3)
plt.plot(trend.year_decimal, trend['temperature trend (C)'], '-')
plt.grid(True)
plt.xlabel('year')
plt.ylabel('monthly average temperature (C)')
plt.title("Temperature trend in Hawai'i")
plt.tight_layout()
plt.savefig('../../manuscript/figures/hawaii-trend.eps')
# Save a png to use in the repository README
plt.savefig('../../manuscript/figures/hawaii-trend.png')
```
